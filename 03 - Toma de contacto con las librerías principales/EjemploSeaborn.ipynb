# Ejemplos de uso de Seaborn

Seaborn es otra librería para representación gráfica de datos, que permite definir otros tipos de gráficos más avanzados o complejos.

Comenzaremos por importar la librería antes de empezar

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Gráficos simples

Podemos construir gráficos sencillos al estilo de Matplotlib, como por ejemplo este gráfico de dispersión donde se muestra la relación entre la longitud de pétalo y sépalo en las distintas especies de la flor "iris", coloreando cada especie de un color

```python
datos = pd.read_csv('Iris.csv')
sns.scatterplot(x='SepalLengthCm', y='PetalLengthCm', hue='Species', data=datos)
plt.show()
```

## Gráficos de enjambre (swarm plots)

Los gráficos de enjambre permiten dividir en categorías horizontales la distribución de valores de un determinado parámetro. En este caso por ejemplo mostramos la longitud de pétalo de las 3 especies diferentes de flor iris.

```python
sns.swarmplot(x='Species', y='PetalLengthCm', data=datos)
plt.show()
```

## Gráficos de regresión

Calculan un diagrama de dispersión (como los *scatter plots*) añadiendo la recta de regresión, recta que mejor se ajusta a los puntos representados.

```python
sns.regplot(x='PetalLengthCm', y='SepalLengthCm', data=datos)
plt.show()
```

## Gráficos de correlación

Establecen en una tabla la correlación (directa o inversa) entre cada pareja de datos de un dataset. Podemos verlo a continuación con los parámetros de precios de venta de inmuebles. Vemos que en el precio de venta influyen principalmente los metros cuadrados de la planta baja, de la primera planta, el número de aseos y el año de construcción.

```python
datos = pd.read_csv('precios_inmuebles.csv')
datos = datos.drop('Barrio', axis=1)
heatmap = sns.heatmap(datos.corr(), vmin=-1, vmax=1, annot=True)
heatmap.set_title('Correlación datos precios inmuebles');
plt.show()
```
