# Ejemplos de gráficos en Matplotlib

En este documento mostraremos cómo construir algunos gráficos habituales en Matplotlib.

Comenzaremos importando la librería

```python
import matplotlib.pyplot as plt
```

## Gráfico de líneas

Ejemplo de gráfico de líneas con valores de X e Y

```python
x = [1, 2, 3, 4, 5]
y = [2, 8, 4, 7, 9]
plt.plot(x, y)
plt.show()
```

## Gráfico de dispersión

Ejemplo de gráfico de dispersión con los mismos valores de X e Y

```python
plt.scatter(x, y)
plt.show()
```

## Gráficos circulares

Ejemplo de gráfico circular que muestra cuántas notas hay de cada tipo

```python
calificaciones = ['Suspensos', 'Aprobados', 'Notables', 'Sobresalientes']
valores = [5, 3, 8, 4]
plt.pie(valores, labels=calificaciones)
plt.show()
```

## Histogramas

Ejemplo de histograma que representa la distribución de alturas de una muestra de individuos

```python
alturas = [185, 170, 185, 190, 175, 190, 185, 175,
    180, 180, 185, 190, 175, 190, 185, 170, 165, 195,
    180, 185, 190, 190, 185, 180]
plt.hist(alturas, bins=7)
plt.show()
```

## Gráficos de barras

Los gráficos de barras muestran correspondencias entre dos secuencias de valores X e Y. Podemos representar más de una correspondencia en un mismo gráfico, con distintos colores para cada una

```python
x = [5, 8, 10]
y = [12, 16, 6]
x2 = [6, 9, 11]
y2 = [6, 15, 7]
plt.bar(x, y, align='center')
plt.bar(x2, y2, color='g', align='center')
plt.title('Gráficos de barras')
plt.ylabel('Eje Y')
plt.xlabel('Eje X')
plt.show()
```

También podemos apilar los gráficos de las secuencias si las coordenadas X de cada una coinciden, usando el parámetro `bottom`

```python
x = [1, 2, 3]
y = [1, 2, 3]
x2 = [1, 2, 3]
y2 = [2, 4, 6]
plt.bar(x, y)
plt.bar(x2, y2, bottom=y, color='r')
plt.show()
```

## Gráficos de cajas

Los gráficos de cajas permiten ver la distribución de un conjunto de valores y detectar gráficamente valores anómalos en el conjunto

```python
x1 = [10, 20, 15, 12, 15, 20, 25, 10, 90, 20, 10, 8, 28, 13, 26, 5]
x2 = [40, 41, 43, 44, 46, 48, 5, 38, 50, 40, 39, 49]
plt.boxplot([x1, x2], tick_labels=["x1", "x2"])
plt.show()
```
