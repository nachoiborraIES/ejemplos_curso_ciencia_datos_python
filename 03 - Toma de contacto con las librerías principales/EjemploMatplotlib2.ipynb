# Opciones adicionales con Matplotlib

En este cuaderno veremos algunas opciones de configuración adicionales que se pueden usar en matplotlib, como el uso de colores, títulos, leyendas, o mallas multi-gráfico.

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Títulos del gráfico y tipos de letra

Podemos especificar un título general para el gráfico y otros específicos para cada eje, e indicar el tipo de letra para cada uno

```python
x = [5, 8, 10]
y = [12, 16, 6]

font1 = {'family': 'DejaVu Sans', 'color': 'red' , 'size': 20}
font2 = {'family': 'Liberation Sans', 'color': 'blue', 'size': 14}

plt.bar(x, y, align='center')
plt.title('Gráficos de barras', fontdict=font1)
plt.ylabel('Eje Y', fontdict=font2)
plt.xlabel('Eje X', fontdict=font2)
plt.show()
```

## Colores, estilos de línea y marcadores

Podemos especificar distintos tipos de colores, como letras simples o en formato hexadecimal. También estilos de línea con `linestyle` y `linewidth`, y marcadores de puntos clave con el parámetro `marker`

```python
x = [1, 2, 3, 4, 5]
y = [2, 8, 4, 7, 9]
plt.plot(x, y, color='#428A92', linestyle='--', linewidth=2, marker='s')
plt.show()
```

## Leyendas

Las leyendas muestran cuadros informativos de qué significa cada parte del gráfico. Podemos especificar la etiqueta (`label`) asociada a cada parte y la ubicación de la leyenda.

```python
x = [1, 2, 3, 4, 5]
y = [2, 8, 4, 7, 9]
x2 = [1, 2, 3, 4, 5]
y2 = [9, 9, 2, 6, 8]
plt.plot(x, y, label='Linea 1', color='r')
plt.plot(x2, y2, label='Linea 2', color='g')
plt.legend(loc='upper center')
plt.show()
```

## Límites de los ejes

Podemos controlar el rango de valores que mostrar en cada eje e incluso los valores a mostrar en cada división intermedia

```python
x = [1, 2, 3, 4, 5]
y = [2, 8, 4, 7, 9]
plt.xlim(1, 5)
plt.xticks(range(1, 6, 1))
plt.plot(x, y, color='#428A92', linestyle='--', linewidth=2, marker='s')
plt.show()
```

## Tamaño del gráfico

Con la instrucción `figure` podemos especificar el tamaño del gráfico (ancho y alto). El tamaño se especifica en pulgadas, por defecto

```python
x = [1, 2, 3, 4, 5]
y = [2, 8, 4, 7, 9]
plt.figure(figsize=(4, 2))
plt.xlim(1, 5)
plt.xticks(range(1, 6, 1))
plt.plot(x, y, color='#428A92', linestyle='--', linewidth=2, marker='s')
plt.show()
```

## Mallas de gráficos

Podemos indicar mallas de gráficos con `subplots` e indicar qué gráfico va en cada casilla

```python
# Datos para el histograma
alturas = [185, 170, 185, 190, 175, 190, 185, 175,
    180, 180, 185, 190, 175, 190, 185, 170, 165, 195,
    180, 185, 190, 190, 185, 180]

# Datos para el gráfico circular
calificaciones = ['Suspensos', 'Aprobados', 'Notables', 'Sobresalientes']
valores = [5, 3, 8, 4]

# Datos para el gráfico de barras
x = [5, 8, 10]
y = [12, 16, 6]
x2 = [6, 9, 11]
y2 = [6, 15, 7]

# Datos para el gráfico de dispersión
x3 = [1, 2, 3, 4, 5]
y3 = [2, 8, 4, 7, 9]

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))
axes[0, 0].hist(alturas, bins=7)
axes[0, 0].set_title('Histograma de alturas')
axes[0, 1].pie(valores, labels=calificaciones, autopct='%1.1f%%')
axes[0, 1].set_title('Gráfico circular')
axes[1, 0].bar(x, y, align='center')
axes[1, 0].bar(x2, y2, align='center', color='g')
axes[1, 0].set_title('Gráfico de barras')
axes[1, 1].scatter(x3, y3)
axes[1, 1].set_title('Gráfico de dispersión')
fig.suptitle("Ejemplo de múltiples gráficos")
plt.show()
```

## Acumular datos en el gráfico

Podemos igualmente utilizar un mismo lienzo de dibujo para representar varios gráficos. Por ejemplo, aquí representamos un histograma con alturas de una muestra de individuos junto con dos líneas verticales (`axvline`) que representan cuál es la media y la mediana de esa muestra.

```python
# Alturas de un grupo de individuos
datos = np.array([185, 170, 180, 192, 178, 181, 185, 195, 168, \
    172, 210, 215, 210, 181, 202, 178])

# Calculamos la media y la mediana de los datos
media = np.mean(datos)
mediana = np.median(datos)

# Creamos el histograma
plt.hist(datos, bins=7, alpha=0.7, color='skyblue', edgecolor='black')

# Agregamos una línea vertical en la media y otra con la mediana
plt.axvline(media, color='red', label=f'Media: {media:.2f}')
plt.axvline(mediana, color='green', label=f'Mediana: {mediana:.2f}')

plt.title('Histograma de alturas')
plt.legend()
plt.show()
```
